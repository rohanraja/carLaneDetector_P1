```python
from numpy import *
import pylab

# data to fit
x = random.rand(6)*255
y = random.rand(6)*255
x = [294, 314, 340, 349, 346, 412, 420]

x = [562, 880, 535, 510, 527, 537]


y = [458, 438, 418, 412, 412, 360, 363]

x = np.array([ 510, 527, 538, 507, 511, 530])
y = np.array([ 326, 335, 340, 320, 315, 307])


y = 540 - np.array(y)

# fit the data with a 4th degree polynomial
z4 = polyfit(x, y, 8 ) 
p4 = poly1d(z4) # construct the polynomial 

z5 = polyfit(x, np.tan(y), 1)
p5 = poly1d(z5)

xx = linspace(0, 960, 100)
pylab.plot(x, y, 'o', xx, p4(xx),'-g', xx, np.tanh(p5(xx)),'-b')
pylab.legend(['data to fit', '4th degree poly', '5th degree poly'])
pylab.axis([0,960,0,540])
pylab.show()
```
